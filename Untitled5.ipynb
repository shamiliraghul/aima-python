```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(
    { "figure.figsize": (6, 4) },
    style='ticks',
    color_codes=True,
    font_scale=0.8
)
%config InlineBackend.figure_format = 'retina'
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.metrics import PredictionErrorDisplay
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
adv = pd.read_csv(
    'https://raw.githubusercontent.com/datacamp/Hacker-Stats-in-Python-Live-Training/master/data/gfmt_sleep.csv',
    index_col=0
)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
adv['gender']=label_encoder.fit_transform(adv['gender'])
adv
```

```python
adv=adv.replace('*',0)
print(adv)
```

```python
adv.fillna(0, inplace=True)
```

```python
adv
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10,3), sharex=True)
sns.scatterplot(data=adv, x='gender', y='sci', alpha=1, ax=axs[0])
sns.scatterplot(data=adv, x='age', y='sci', alpha=1, ax=axs[1]);
```

```python
X = adv.drop(columns='sci')
y = adv['sci']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
X_age = X[['age']]
```

```python
X_age_train, X_age_test = X_train[['age']], X_test[['age']]
```

```python
lr1d = LinearRegression()
```

```python
lr1d.fit(X_age_train, y_train)
```

```python
a_linear_test_set = pd.DataFrame(
    np.linspace(X[['age']].min(), X[['age']].max(), 500), columns=X_age_train.columns
)
```

```python
y_pred = lr1d.predict(a_linear_test_set)
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.scatter(X_age_train['age'], y_train, alpha=0.7)
ax.plot(a_linear_test_set, y_pred, '-k', alpha=0.7);
```

```python
mean_absolute_error(y_train, lr1d.predict(X_train[['age']]))
```

```python
mean_absolute_error(y_test, lr1d.predict(X_test[['age']]))
```
